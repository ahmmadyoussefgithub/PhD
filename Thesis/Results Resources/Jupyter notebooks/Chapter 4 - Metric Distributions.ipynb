```python
# imports and data read

import seaborn as sns
import numpy as np
from scipy import stats
import pandas 

data = pandas.read_csv('/Volumes/Windows HD/PhD/team_size.tsv','\t')
metrics = ['CBO', 'DIT', 'LCOM', 'NOC', 'RFC', 'WMC']
```

```python
# plot distributions for each metric

for metric in metrics:
    distplot = sns.distplot(data[metric].astype(int), hist=False);
    sns.plt.show()
```

```python
# Get the underlying data to plot in Excel

for metric in metrics:
    print('---------->', metric)
    distplot = sns.distplot(data[metric].astype(int), hist=False);
    for i in range (0, len(distplot.get_lines()[0].get_data()[1])):
        print(distplot.get_lines()[0].get_data()[0][i], '	', distplot.get_lines()[0].get_data()[1][i])
```
