```python
#https://plot.ly/~AhmedYoussef

import matplotlib.pyplot
import plotly.plotly 
import plotly.graph_objs
from plotly import figure_factory 

import numpy
import pandas 
import scipy.stats
import statsmodels.api

data = pandas.read_csv('/Volumes/Windows HD/PhD/team_size.tsv','\t')
metrics = ['CBO', 'DIT', 'LCOM', 'NOC', 'RFC', 'WMC']
```

```python
#https://matplotlib.org/examples/pylab_examples/boxplot_demo.html

for metric in metrics:
    revision_groups = [data[(data['Class Revision Count']==1)][metric].values, 
                       data[(data['Class Revision Count']==2)][metric].values,
                       data[(data['Class Revision Count']==3)][metric].values,
                       data[(data['Class Revision Count']==4)][metric].values,
                       data[(data['Class Revision Count']==5)][metric].values,
                       data[(data['Class Revision Count']==6)][metric].values]
    matplotlib.pyplot.boxplot(revision_groups, sym='')
    matplotlib.pyplot.title(metric, fontsize=14, fontweight='bold')
    matplotlib.pyplot.show()
```
