```python
import seaborn as sns
import numpy as np
from scipy import stats
import pandas
import statsmodels.api

data = pandas.read_csv('/Volumes/Windows HD/PhD/team_stability_interproject.tsv', delimiter='\t')

metrics = ['CBO', 'DIT', 'LCOM', 'NOC', 'RFC', 'WMC']
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

x=(data[['Categorical Lack of Stability']])
variance_inflation_factor(exog=x.values, exog_idx=0)
```

```python

for metric in metrics:
    model = statsmodels.api.OLS(data[metric], data['Categorical Lack of Stability'])
    result = model.fit()
    print(result.summary())
```
