```python
import seaborn as sns
import numpy as np
from scipy import stats
import pandas
import statsmodels.api
from matplotlib import pyplot as plt


data = pandas.read_csv('/Volumes/Windows HD/PhD/team_stability.tsv', delimiter='\t')
```

```python
# correlation_matrix = np.corrcoef([data['CBO'], data['DIT'], data['NOC'], data['LCOM'], data['RFC'], data['WMC'], data['Class Revision Count'], data['Project Author Count'], data['Class Stability']])
# correlation_matrix = np.corrcoef([data['CBO'], data['DIT'], data['NOC'], data['LCOM'], data['RFC'], data['WMC'], data['Class Revision Count'], data['Project Author Count'], data['Class Stability V2']])
correlation_matrix = np.corrcoef([data['Class Revision Count'],  data['Project Author Count'], data['Project Duration'], data['Module Count'], (data['Project Order']-1)*(data['Project Order']-1)])
print(correlation_matrix)
```
